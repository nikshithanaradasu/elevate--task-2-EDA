# Task 2: Exploratory Data Analysis (EDA)

#### Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

%matplotlib inline
sns.set(style="whitegrid")
```

#### Step 2: Load the Cleaned Dataset

```python
df = pd.read_csv('cleaned_titanic.csv')
df.head()
```

#### Step 3: Summary Statistics

```python
# Summary stats
df.describe().T

# Check for nulls again
df.isnull().sum()
```

#### Step 4: Histograms & Boxplots

```python
# Histogram of numeric features
df.hist(figsize=(12,8), bins=20)
plt.suptitle("Histograms of Numeric Features")

# Boxplots
plt.figure(figsize=(10,5))
sns.boxplot(data=df[['Age', 'Fare']])
plt.title("Boxplots of Age and Fare")
```

#### Step 5: Correlation & Pairplot

```python
# Only include numeric columns for correlation
numeric_df = df.select_dtypes(include=['number'])

# Plot correlation matrix
plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

# Pairplot
sns.pairplot(df[['Survived', 'Pclass', 'Sex', 'Age', 'Fare']])
```

#### Step 6: Insights, Patterns & Anomalies

```python
# Survival by Sex
sns.barplot(x='Sex', y='Survived', data=df)
plt.title("Survival Rate by Gender")

# Survival by Class
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title("Survival Rate by Passenger Class")
```

#### Step 7: Final Observations

- Females had higher survival rate than males.
- 1st class passengers survived more often.
- Age and Fare are skewed.
- 'Sex' and 'Pclass' are strong predictors.
